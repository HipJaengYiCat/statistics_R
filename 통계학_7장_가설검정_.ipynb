#z검정 & t검정
모평균의 검정을 표준정규분포를 따르는 1만개의 가상데이터를 바탕으로 검정실시

```
#패키지
install.packages("mosaic")
install.packages("TeachingDemos")
library(mosaic)
library(TeachingDemos)
```

```
#가상데이터 생성
n_num = 10000
```

##모분산 아는 경우 z통계량으로 모평균이 0인 귀무가설 검정

```
#z_test
z_test = do(n_num)*z.test(rnorm(10), stdev = 1)$statistic
tally(~(abs(z)>qnorm(0.975)), data = z_test)/n_num
```

```
#z_test histogram
histogram(~z, data = z_test, width = 1, group = abs(z) > qnorm(0.975), fit = "normal", xlim = c(-5, 5), breaks = c(-20:20)/2.5, type = "density")
```

##모분산을 모르는 경우 t검정 통계량으로 모평균이 0인 귀무가설 검정

```
#t_test
t_test = do(n_num)*t.test(rnorm(10))$statistic
tally(~(abs(t)>qnorm(0.975)), data = t_test)/n_num
```

```
#t_test histogram
histogram(~t, data = t_test, width = 1, groups = abs(t)>qt(0.975, 9), fit = "t", xlim = c(-5, 5), breaks = c(-20:20)/2, type = "density")
```
