# 4. 코로나19관련 한국, 중국, 미국, 프랑스의 코로나 19 신규 확진자 수의 추이를 데이터시각화로 비교, 향후 추이에 대해 의견 기술

```
data = 'https://raw.githubusercontent.com/seunghee0518/R_File/main/data.csv'
data = read.csv(data, header = T)
head(data)
```

```
data.KR = subset(data, data$geoId == 'KR')
data.CN = subset(data, data$geoId == 'CN')
data.US = subset(data, data$geoId == 'US')
data.FR = subset(data, data$geoId == 'FR')
```

```
head(data.US[, 1], 3)
tail(data.US[, 1], 3)
```

```
data.FR[, 5][195]
```

```
#잘못된 값 수정
data.FR[, 5][195] = 766
data.FR[, 5][195]
```

```
install.packages('zoo')
library(zoo)
```

```
#한국데이터
date = seq(as.Date('2019-12-31'), as.Date('2020-12-14'), 'day')
case.KR = rev(data.KR[, 5])
case.KR = zoo(case.KR, date)
plot(case.KR)
```

```
#중국
case.CN = rev(data.CN[, 5])
case.CN = zoo(case.CN, date)
plot(case.CN)
```

```
#미국
case.US = rev(data.US[, 5])
case.US = zoo(case.US, date)
plot(case.US)
```

```
#프랑스
case.FR = rev(data.FR[, 5])
case.FR = zoo(case.FR, date)
plot(case.FR)
```
